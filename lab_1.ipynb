```python
def find_month(x):
    if " 01 " in x:
        return "Jan"
    elif " 02 " in x:
        return "Feb"
    elif " 03 " in x:
        return "March"    
    elif " 04 " in x:
        return "April"    
    elif " 05 " in x:
        return "May"    
    elif " 06 " in x:
        return "June"    
    elif " 07 " in x:
        return "July"    
    elif " 08 " in x:
        return "August"    
    elif " 09 " in x:
        return "Sep"    
    elif " 10 " in x:
        return "Oct"    
    elif " 11 " in x:
        return "Nov"    
    else:
        return "Dec"
```

```python
df["Month"]=df["Date/Time"].apply(find_month)
```

```python
df["Loss"] = df["Theoretical_Power_Curve (KWh)"]- df["LV ActivePower (kW)"] 
df.head()
```

```python
df["Loss_percentage"] = (df["LV ActivePower (kW)"] / df["Theoretical_Power_Curve (KWh)"]) * 100
df.head()
```

```python
# These new loss column represents how much power is lost between the theoretical power calculations to the actual power that 
# was recorded. The first loss column provides the actual value of loss while the loss_percentage column; can emulate the percent yield, or the percent error calculation. 
# That column displays the effectiveness of the data, and how much power was actually gained compared to the maximum power that was recorded/ calculated. 
```

```python
x = df["Wind Speed (m/s)"]
y = df["Loss"]
plt.scatter(x,y)
```

```python
# Looking at the scatter plot of wind speed and loss; there is no correlation between the two values. In the scatterplot there isn't 
# any define trend line between wind speed and loss, and wind speed and loss percentage. Therefore, you could use this data to infer that 
# the wind speed will not cause an increase or decrease in loss of power. And that it has no relation between the loss percentage between actual and theoretical power. 
```

```python
x = df["Wind Speed (m/s)"]
y = df['Loss_percentage']
plt.scatter(x,y)
```

```python
# Something that could be creating this loss, is external factors. When calculating the theoretical power, you are calculating the amount of 
# energy that the wind turbine should generate. External factors like temperature, humidity, pressure, and precipiation could 
# affect the wind turbines ability to generate a consistent amount of energy, therefore affecting the actual measurements compared to the theoretical power. 
# As previously mentioned before, there isn't any correlation between the wind speed and the loss of power. Therefore, you could predict that the cause of loss has to deal with 
# external factors besides the wind speed. 
# You could try to overcome this loss, by taking in some of these external factors like pressure and temperature into calculating the theoretical power. 
```
